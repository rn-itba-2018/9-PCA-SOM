```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math

N=500

plt.rcParams['figure.figsize'] = (12.0, 10.0)

x1=np.random.rand(N,2)
x2=np.random.rand(N,2)
x3=np.random.rand(N,2)

x2[:,0]+=1;
x3[:,1]+=1;

x=np.concatenate((x1,x2,x3),axis=0)

plt.plot(x[:,0],x[:,1],'.')
plt.show()
```

```python
#data = np.matrix.transpose(data)

from minisom import MiniSom
### Initialization and training ###

som2 = MiniSom(100,1,2,sigma=3,learning_rate=0.5)

som2.random_weights_init(x)

som2.train_random(x,10000)


#print(som2.weights)
#print(som.weights[:,:,0].reshape(-1))
#print(som.weights[:,:,1].reshape(-1))

plt.plot(x[:,0],x[:,1],'.')
wx =som2.weights[:,:,0].reshape(-1)
wy = som2.weights[:,:,1].reshape(-1)
plt.plot(wx,wy,'o-')
plt.show()
```
