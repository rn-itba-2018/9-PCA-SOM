```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math

plt.rcParams['figure.figsize'] = (10.0, 8.0)

fr=0.1
N=20


a=np.random.rand(N)

b=np.random.rand(N)

x = np.column_stack((a,b))



#Grafico x1 y x2 
plt.plot(x[:,0],x[:,1],'o')
plt.show()
```

```python
#data = np.matrix.transpose(data)

from minisom import MiniSom
### Initialization and training ###

som2 = MiniSom(150,1,2,sigma=6,learning_rate=0.5)

som2.random_weights_init(x)

som2.train_random(x,10000)

wx =som2.weights[:,:,0].reshape(-1)
wy = som2.weights[:,:,1].reshape(-1)

plt.scatter(x[:,0],x[:,1],s=120,c='r')
plt.plot(wx,wy,'.-',c='b')
plt.show()
```
