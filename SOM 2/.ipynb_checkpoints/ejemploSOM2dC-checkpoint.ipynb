```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
import sklearn.datasets as ds

fr=0.1
N=500
factor=0.3

x = np.random.rand(N,2)

#Grafico x

plt.scatter(x[:,0], x[:,1], s=10)
plt.show()
```

```python
#data = np.matrix.transpose(data)

from minisom import MiniSom
### Initialization and training ###

som2 = MiniSom(10,10,2,sigma=0.5,learning_rate=0.5)

som2.random_weights_init(x)

som2.train_batch(x,100)


#print(som2.weights)
#print(som.weights[:,:,0].reshape(-1))
#print(som.weights[:,:,1].reshape(-1))

wx =som2.weights[:,:,0].reshape(-1)
wy = som2.weights[:,:,1].reshape(-1)

plt.scatter(x[:,0],x[:,1], s=10)
#plt.scatter(wx,wy,s=40)

plt.plot(wx,wy,'or')
plt.show()
```

```python
from pylab import plot,axis,show,pcolor,colorbar,bone

markers = ['o','s']
colors = ['r','g']

for cnt,xx in enumerate(x):
 w = som2.winner(xx) # getting the winner
 # palce a marker on the winning position for the sample xx
 plot(w[0]+.5,w[1]+.5,'o',markerfacecolor='None',
   markeredgecolor='r',markersize=12,markeredgewidth=2)
axis([0,som2.weights.shape[0],0,som2.weights.shape[1]])
show()
```

```python
qnt = som2.quantization(x)
plt.plot(qnt[:,0],qnt[:,1],'x')
plt.show()

```
