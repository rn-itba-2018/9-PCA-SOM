```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
import sklearn.datasets as ds

plt.rcParams['figure.figsize'] = (12.0, 10.0)

fr=0.1
N=500
factor=0.3

x = np.random.rand(N,2)

#Grafico x

plt.scatter(x[:,0], x[:,1], s=10)
plt.show()
```

```python
#data = np.matrix.transpose(data)

from minisom import MiniSom
### Initialization and training ###

X=10
Y=10
som2 = MiniSom(X,Y,2,sigma=1.5,learning_rate=0.2)

som2.random_weights_init(x)

som2.train_random(x,10000)

for i in range(0,X):
    plt.plot(som2.weights[i,:,0],som2.weights[i,:,1],'o-',c='k')
for i in range(0,Y):
    plt.plot(som2.weights[:,i,0],som2.weights[:,i,1],'o-',c='k')

plt.scatter(x[:,0],x[:,1], s=10)
plt.show()
```

```python
from pylab import plot,axis,show,pcolor,colorbar,bone

markers = ['o','s']
colors = ['r','g']

for cnt,xx in enumerate(x):
 w = som2.winner(xx) # getting the winner
 # palce a marker on the winning position for the sample xx
 plot(w[0]+.5+np.random.randn(1)/10,w[1]+.5+np.random.randn(1)/10,'o',markerfacecolor='None',
   markeredgecolor='r',markersize=2,markeredgewidth=2)
axis([0,som2.weights.shape[0],0,som2.weights.shape[1]])
show()
```
