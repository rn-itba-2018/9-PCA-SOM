```python
import wfdb
import numpy as np
import os
import matplotlib.pyplot as plot
from IPython.display import display


plot.rcParams['figure.figsize'] = (12.0, 10.0)
```

```python
# Importo los datos del dataset 100 de mitdb
signals2, fields2=wfdb.srdsamp('114', channels=[0,1], sampfrom=0, sampto=600000, pbdir = 'mitdb') 

# Importo las anotaciones del mismo dataset
annotation = wfdb.rdann('114', 'atr', sampfrom = 0, sampto = 600000, pbdir = 'mitdb')

#Me quedo con los picos de cada ECG que es donde están las anotaciones
picos=annotation.annsamp
#print(annotation.anntype.index("V"))
#Transpongo las señales leídas para poder operar mas comodamente
signals=np.matrix.transpose(signals2)

picos[-1]=0 #descarto el último pico por si el latido está interrumpido
picos[-2]=0 #descarto el último pico por si el latido está interrumpido
picos[0]=0 #descarto el último pico por si el latido está interrumpido
picos[1]=0 #descarto el último pico por si el latido está interrumpido


tam=125 #Tamaño del latido que tomo a la izquierda y a la derecha de la anotación

cantLatidos=sum(picos>tam) #Cuento la cantidad de anotaciones en el segemento a analizar

latidos=np.zeros([cantLatidos,2*tam]) #Reservo espacio para la matriz de datos cantLatidos x 2*tam

latidosNorm=np.zeros([2*tam,cantLatidos])#Reservo espacio para la matriz de datos normalizados

idxLat=0 #Inicializo en cero para procesar cada uno de los picos

for i in picos:
    if i>tam: # i>tam por si el primer latido está trunco
        latidos[idxLat]=signals[1][i-tam:i+tam] #tomo tam muestras a la izquerda y a la derecha de la anotación
        idxLat+=1 #incremento el número de latidos

#El set de latidos está en latidos y el largo de cada latido es de tam muestras

latidos=np.matrix.transpose(latidos) #vuelvo a acondicionar los latidos para seguir operando cómodamente

# A continuación calculo la media y la varianza de cada elemento del vector de datos para poder normalizar cada dato

medias=np.zeros(2*tam)
desv=np.zeros(2*tam)
for i in range(0,2*tam):
    medias[i]=np.mean(latidos[i,:])
    desv[i]=np.std(latidos[i,:])
```

```python
kinds = set(annotation.anntype)
print(kinds)

for i in kinds:
    print(i)
    print(annotation.anntype.count(i))
    #print(annotation.anntype.index(i))
    
indexesA = [i-2 for i,x in enumerate(annotation.anntype) if x == 'A']
indexesN = [i-2 for i,x in enumerate(annotation.anntype) if x == 'N']
indexesV = [i-2 for i,x in enumerate(annotation.anntype) if x == 'V']
indexesJ = [i-2 for i,x in enumerate(annotation.anntype) if x == 'J']
indexesF = [i-2 for i,x in enumerate(annotation.anntype) if x == 'F']

print(indexesA)

indexesN = indexesN[0:-2]

latidosA = latidos[:,indexesA]
latidosV = latidos[:,indexesV]
latidosJ = latidos[:,indexesJ]
latidosF = latidos[:,indexesF]
latidosN = latidos[:,indexesN]


trainA = latidosA[:,range(0,6)]
trainN = latidosN[:,range(0,6)]
trainV = latidosV[:,range(0,6)]
trainJ = latidosJ
trainF = latidosF

latidostrain = np.concatenate((trainA,trainN,trainV,trainJ,trainF),axis=1)
```

```python

plot.plot(trainA)
plot.show()
plot.plot(trainN)
plot.show()
plot.plot(trainV)
plot.show()
plot.plot(trainJ)
plot.show()
plot.plot(trainF)
plot.show()
```

```python
#Grafico los latidos normalizados
plot.plot(latidostrain)
plot.show()


latidostrain = np.transpose(latidostrain)
```

```python

from minisom import MiniSom
from matplotlib import gridspec

### Initialization and training ###

X=8
Y=8

som2 = MiniSom(X,Y,250,sigma=5,learning_rate=1)

som2.random_weights_init(latidostrain)

som2.train_random(latidostrain,2000)

fig = plot.figure()

gs = gridspec.GridSpec(X,Y)

for i in range(0,X):
    for j in range(0,Y):
        ax1 = fig.add_subplot(gs[i,j])
        ax1.plot(som2.weights[i,j,:],'-',c='k')

        
gs.update(wspace=0.5, hspace=0.5)
fig = plot.gcf()

plot.show()
```
